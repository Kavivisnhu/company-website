```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1
```

```python
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
```

```python
df['sepal.width']=np.where(df['sepal.width']>upper_bound['sepal.width'],upper_bound['sepal.width'] ,df['sepal.width'])
df['sepal.width']=np.where(df['sepal.width']<lower_bound['sepal.width'],lower_bound['sepal.width'],df['sepal.width'])
```

```python
print(f"after column capped between: {lower_bound} and{upper_bound}.")
print('after capping:')
print(df['sepal.width'].describe())
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(df['sepal.width'])
plt.title('Boxplot of sepal.width')
plt.ylabel('sepal.width')
plt.show()
```

```python
x=df.drop('variety',axis=1)
y=df['variety']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
x_train,y_train=smote.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```
